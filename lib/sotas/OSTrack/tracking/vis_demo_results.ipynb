```python
seq = 'video_suidao_RGB_ostrack'
rgbs_gt_root = '/media/group2/data/zongtaobu/Demodata/suidao_RGB'
rgbs_results = '/home/test4/code/EventBenchmark/output/test/tracking_results/ostrack/vitb_256_mae_ce_32x4_ep300/'+seq+'.txt'

rgbs_predict = np.loadtxt(rgbs_results)
pic_list = os.listdir(rgbs_gt_root)
pic_list.sort(key = lambda  x:int(x[:-4]))
i = 0
img = os.path.join(rgbs_gt_root,pic_list[i])
img = cv2.imread(img)
box = rgbs_predict[i,:]
rect3 = patches.Rectangle((box[0], box[1]), box[2], box[3],fill=False, edgecolor='red', linewidth=1)
fig,ax = plt.subplots()
ax.add_patch(rect3)
ax.imshow(img)
ax.set_axis_off()
plt.show()


img0 = cv2.imread(os.path.join(rgbs_gt_root,pic_list[0]))
h,w = img0.shape[:2]

video_path='/home/test4/code/EventBenchmark/output/test/result_plots/'+seq
os.makedirs(video_path,exist_ok=True)
# videoWriter = cv2.VideoWriter(video_path, fourcc, fps, (h,w)) #
# for i in range(2100,2600,10):
for i in range(rgbs_predict.shape[0]):
    img = os.path.join(rgbs_gt_root,pic_list[i])
    img = cv2.imread(img)
    box = rgbs_predict[i,:]
    rect3 = patches.Rectangle((box[0], box[1]), box[2], box[3],fill=False, edgecolor='red', linewidth=1)
    fig,ax = plt.subplots()
    ax.add_patch(rect3)
    ax.imshow(img)
    ax.set_axis_off()
    plt.savefig(video_path+'/{}.jpg'.format(i),bbox_inches='tight', pad_inches = -0.1)
    plt.close()
```
